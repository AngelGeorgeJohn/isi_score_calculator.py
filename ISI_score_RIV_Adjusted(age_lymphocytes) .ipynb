## Merge with encounter frame

```python
import numpy as np
import pandas as pd

def merge_visits_with_mtp(visits, mtp_wide):

    df = visits.copy()
    mtp = mtp_wide.copy()

    # Ensure Date_Of_Visit is datetime
    df["Date_Of_Visit"] = pd.to_datetime(df["Date_Of_Visit"], errors="coerce")

    # Merge wide IV data onto visits
    df = df.merge(mtp, on="RKD.ID", how="left")

    # Identify dose blocks
    dose_numbers = sorted(
        int(col.split("_")[1])
        for col in mtp.columns
        if col.startswith("mtp_") and col.endswith("_dose_date")
    )

    for n in dose_numbers:

        dose_date_col = f"mtp_{n}_dose_date"
        value_col     = f"mtp_{n}_dose_value"

        # Convert IV dates to datetime
        df[dose_date_col] = pd.to_datetime(df[dose_date_col], errors="coerce")

        # Condition:
        # Keep the dose if the visit occurs AFTER the dose date
        # Blank it otherwise
        mask = (
            df[dose_date_col].notna() &
            (df[dose_date_col] < df["Date_Of_Visit"])
        )

        # Blank outside window
        df.loc[~mask, [dose_date_col, value_col]] = [np.nan, np.nan]

    return df
```

## Reorder columns

```python
def reorder_mtp_columns(df):

    cols = list(df.columns)

    # Identify dose numbers based ONLY on dose_date
    dose_numbers = sorted(
        int(c.split("_")[1])
        for c in cols
        if c.startswith("mtp_") and c.endswith("_dose_date")
    )

    # 1) Core columns (non-mtp)
    core_cols = [c for c in cols if not c.startswith("mtp_")]

    # 2) Dose blocks (ordered)
    dose_blocks = []
    for n in dose_numbers:
        block = [
            f"mtp_{n}_dose_date",
            f"mtp_{n}_dose_value",
            f"mtp_{n}_dose_unit",
            f"mtp_{n}_vanish_days",
            f"mtp_{n}_vanish_date",
        ]
        dose_blocks.extend([c for c in block if c in cols])

    # 3) Everything else starting with mtp_ (course columns, intervals, etc.)
    other_mtp_cols = [
        c for c in cols
        if c.startswith("mtp_") and c not in dose_blocks
    ]

    # Build final ordering
    ordered_cols = core_cols + dose_blocks + other_mtp_cols

    return df[ordered_cols]

```

```python
import numpy as np
import pandas as pd

def min_interval_table(df, max_dose=14):
    x = df.copy()

    # Parse dose dates
    for n in range(1, max_dose + 1):
        col = f"mtp_{n}_dose_date"
        if col in x.columns:
            x[col] = pd.to_datetime(x[col], errors="coerce", dayfirst=True)

    # Use earliest dose dates per patient (avoids repeated visit rows)
    per_patient = (
        x.groupby("RKD.ID")[[c for c in x.columns if c.startswith("mtp_") and c.endswith("_dose_date")]]
         .min()
         .reset_index()
    )

    # Build min interval per dose (global min across dataset)
    rows = []
    d1 = "mtp_1_dose_date"

    for n in range(2, max_dose + 1):
        dn = f"mtp_{n}_dose_date"
        if dn not in per_patient.columns:
            continue

        interval_series = (per_patient[dn] - per_patient[d1]).dt.days

        rows.append({
            "interval_name": f"interval_{n}_minus_1_days",
            "min_days": interval_series.min(skipna=True)
        })

    return pd.DataFrame(rows)

# Example:
min_intervals_table = min_interval_table(mtp_merged_data, max_dose=14)
min_intervals_table
```

```python
def update_mtp_doses(df, base_dose, target_dose, max_days=30):
    result = df.copy()

    base_date   = f"mtp_{base_dose}_dose_date"
    target_date = f"mtp_{target_dose}_dose_date"
    base_value  = f"mtp_{base_dose}_dose_value"
    target_value= f"mtp_{target_dose}_dose_value"

    # Parse dates
    result[base_date]   = pd.to_datetime(result[base_date], errors="coerce", dayfirst=True)
    result[target_date] = pd.to_datetime(result[target_date], errors="coerce", dayfirst=True)

    # Parse numeric values
    result[base_value]   = pd.to_numeric(result[base_value], errors="coerce")
    result[target_value] = pd.to_numeric(result[target_value], errors="coerce")

    def _per_patient(g):
        g = g.copy()

        d_base   = g[base_date].dropna()
        d_target = g[target_date].dropna()

        # Need both doses
        if d_base.empty or d_target.empty:
            return g

        base_date_val   = d_base.iloc[0]
        target_date_val = d_target.iloc[0]

        # Check interval
        interval_days = (target_date_val - base_date_val).days
        if pd.isna(interval_days) or interval_days > max_days:
            return g

        # Get dose values
        v_base   = g.loc[g[base_date].notna(), base_value].dropna()
        v_target = g.loc[g[target_date].notna(), target_value].dropna()

        base_val   = v_base.iloc[0]   if not v_base.empty   else np.nan
        target_val = v_target.iloc[0] if not v_target.empty else np.nan

        cumulative = base_val + target_val

        # Replace ALL non-missing base-dose rows
        mask = g[base_date].notna()
        g.loc[mask, base_date]  = target_date_val
        g.loc[mask, base_value] = cumulative

        # Blank target-dose after merge
        g[target_date]  = pd.NaT
        g[target_value] = np.nan

        return g

    return result.groupby("RKD.ID", group_keys=False).apply(_per_patient)
```

```python
import numpy as np
import pandas as pd

# ---------------------------------------------
# 1. Sigmoid function (UNCHANGED)
# ---------------------------------------------
def sigmoid_curve(x, A, n, d):
    return A / (1 + np.exp(n * (x - d)))

# ---------------------------------------------
# 2. Interpolation functions for A(dose), d(dose), n(dose) (UNCHANGED)
# ---------------------------------------------
A1 = 0.40
A2 = 0.80
dose1 = 250
dose2 = 3000

m_A = (A2 - A1) / (dose2 - dose1)
b_A = A1 - m_A * dose1
def calculate_A(dose):
    return m_A * dose + b_A

d1 = 15
d2 = 21
m_d = (d2 - d1) / (dose2 - dose1)
b_d = d1 - m_d * dose1
def calculate_d(dose):
    return m_d * dose + b_d

# vanish logic (UNCHANGED)
def calculate_n_from_vanish(A, d, vanish_day):
    return np.log(99) / (vanish_day - d)

vanish1 = 21
vanish2 = 30
n1 = calculate_n_from_vanish(A1, d1, vanish1)
n2 = calculate_n_from_vanish(A2, d2, vanish2)

m_n = (n2 - n1) / (dose2 - dose1)
b_n = n1 - m_n * dose1
def calculate_n(dose):
    return m_n * dose + b_n

# ---------------------------------------------
# 3. ITIS computation for ALL doses 1–11 (equations unchanged)
# ---------------------------------------------
def compute_ITIS_generic(row, dose_num):
    dose_col = f"mtp_{dose_num}_dose_value"
    time_col = f"days_visit_minus_mtp{dose_num}"

    dose = row[dose_col]
    time = row[time_col]

    # Skip invalid values
    if pd.isna(dose) or pd.isna(time):
        return np.nan

    A = calculate_A(dose)
    d = calculate_d(dose)
    n = calculate_n(dose)

    return sigmoid_curve(time, A, n, d)

# Create ITIS columns for doses 1–11
for n in range(1, 7):
    dose_col = f"mtp_{n}_dose_value"
    time_col = f"days_visit_minus_mtp{n}"
    out_col  = f"mtp_dose{n}_ITIS"

    if dose_col in mtp_merged_data.columns and time_col in mtp_merged_data.columns:
        mtp_merged_data[out_col] = mtp_merged_data.apply(
            lambda row: compute_ITIS_generic(row, n),
            axis=1
        )
```

```python
import numpy as np
import pandas as pd

def merge_visits_with_cyc(visits, cyc_wide):

    df = visits.copy()
    cyc = cyc_wide.copy()

    # Ensure Date_Of_Visit is datetime
    df["Date_Of_Visit"] = pd.to_datetime(df["Date_Of_Visit"], errors="coerce")

    # Merge wide IV data onto visits
    df = df.merge(cyc, on="RKD.ID", how="left")

    # Identify dose blocks
    dose_numbers = sorted(
        int(col.split("_")[1])
        for col in cyc.columns
        if col.startswith("cyc_") and col.endswith("_dose_date")
    )

    for n in dose_numbers:

        dose_date_col = f"cyc_{n}_dose_date"
        value_col     = f"cyc_{n}_dose_value"
     

        # Convert IV dates to datetime
        df[dose_date_col] = pd.to_datetime(df[dose_date_col], errors="coerce")

        # Condition:
        # Keep the dose if the visit occurs AFTER the dose date
        # Blank it otherwise
        mask = (
            df[dose_date_col].notna() &
            (df[dose_date_col] < df["Date_Of_Visit"])
        )

        # Blank outside window
        df.loc[~mask, [dose_date_col, value_col]] = [np.nan, np.nan]

    return df
```

```python
def reorder_cyc_columns(df):

    cols = list(df.columns)

    # Identify dose numbers based ONLY on dose_date
    dose_numbers = sorted(
        int(c.split("_")[1])
        for c in cols
        if c.startswith("cyc_") and c.endswith("_dose_date")
    )

    # 1) Core columns (non-cyc)
    core_cols = [c for c in cols if not c.startswith("cyc_")]

    # 2) Dose blocks (ordered)
    dose_blocks = []
    for n in dose_numbers:
        block = [
            f"cyc_{n}_dose_date",
            f"cyc_{n}_dose_value",
            f"cyc_{n}_vanish_days",
            f"cyc_{n}_vanish_date",
        ]
        dose_blocks.extend([c for c in block if c in cols])

    # 3) Everything else starting with cyc_ (course columns, intervals, etc.)
    other_cyc_cols = [
        c for c in cols
        if c.startswith("cyc_") and c not in dose_blocks
    ]

    # Build final ordering
    ordered_cols = core_cols + dose_blocks + other_cyc_cols

    return df[ordered_cols]

```

```python
def update_cyc_doses(df, base_dose, target_dose, max_days=180):
    result = df.copy()

    base_date   = f"cyc_{base_dose}_dose_date"
    target_date = f"cyc_{target_dose}_dose_date"
    base_value  = f"cyc_{base_dose}_dose_value"
    target_value= f"cyc_{target_dose}_dose_value"

    # Parse dates
    result[base_date]   = pd.to_datetime(result[base_date], errors="coerce", dayfirst=True)
    result[target_date] = pd.to_datetime(result[target_date], errors="coerce", dayfirst=True)

    # Parse numeric values
    result[base_value]   = pd.to_numeric(result[base_value], errors="coerce")
    result[target_value] = pd.to_numeric(result[target_value], errors="coerce")

    def _per_patient(g):
        g = g.copy()

        d_base   = g[base_date].dropna()
        d_target = g[target_date].dropna()

        # Need both doses
        if d_base.empty or d_target.empty:
            return g

        base_date_val   = d_base.iloc[0]
        target_date_val = d_target.iloc[0]

        # Check interval
        interval_days = (target_date_val - base_date_val).days
        if pd.isna(interval_days) or interval_days > max_days:
            return g

        # Get dose values
        v_base   = g.loc[g[base_date].notna(), base_value].dropna()
        v_target = g.loc[g[target_date].notna(), target_value].dropna()

        base_val   = v_base.iloc[0]   if not v_base.empty   else np.nan
        target_val = v_target.iloc[0] if not v_target.empty else np.nan

        cumulative = base_val + target_val

        # Replace ALL non-missing base-dose rows
        mask = g[base_date].notna()
        g.loc[mask, base_date]  = target_date_val
        g.loc[mask, base_value] = cumulative

        # Blank target-dose after merge
        g[target_date]  = pd.NaT
        g[target_value] = np.nan

        return g

    return result.groupby("RKD.ID", group_keys=False).apply(_per_patient)
```

```python
import numpy as np
import pandas as pd

# ---------------------------------------------
# 1. Sigmoid function (UNCHANGED)
# ---------------------------------------------
def sigmoid_curve(x, A, n, d):
    return A / (1 + np.exp(n * (x - d)))

# ---------------------------------------------
# 2. Interpolation functions for A(dose), d(dose), n(dose) (UNCHANGED)
# ---------------------------------------------
A1 = 0.60
A2 = 0.90
dose1 = 150
dose2 = 8000

m_A = (A2 - A1) / (dose2 - dose1)
b_A = A1 - m_A * dose1
def calculate_A(dose):
    return m_A * dose + b_A

d1 = 40
d2 = 80
m_d = (d2 - d1) / (dose2 - dose1)
b_d = d1 - m_d * dose1
def calculate_d(dose):
    return m_d * dose + b_d

# vanish logic (UNCHANGED)
def calculate_n_from_vanish(A, d, vanish_day):
    return np.log(99) / (vanish_day - d)

vanish1 = 60
vanish2 = 110
n1 = calculate_n_from_vanish(A1, d1, vanish1)
n2 = calculate_n_from_vanish(A2, d2, vanish2)

m_n = (n2 - n1) / (dose2 - dose1)
b_n = n1 - m_n * dose1
def calculate_n(dose):
    return m_n * dose + b_n

# ---------------------------------------------
# 3. ITIS computation for ALL doses 1–11 (equations unchanged)
# ---------------------------------------------
def compute_ITIS_generic(row, dose_num):
    dose_col = f"cyc_{dose_num}_dose_value"
    time_col = f"days_visit_minus_cyc{dose_num}"

    dose = row[dose_col]
    time = row[time_col]

    # Skip invalid values
    if pd.isna(dose) or pd.isna(time):
        return np.nan

    A = calculate_A(dose)
    d = calculate_d(dose)
    n = calculate_n(dose)

    return sigmoid_curve(time, A, n, d)

# Create ITIS columns for doses 1–11
for n in range(1, 12):
    dose_col = f"cyc_{n}_dose_value"
    time_col = f"days_visit_minus_cyc{n}"
    out_col  = f"cyc_dose{n}_ITIS"

    if dose_col in cyc_merged_data.columns and time_col in cyc_merged_data.columns:
        cyc_merged_data[out_col] = cyc_merged_data.apply(
            lambda row: compute_ITIS_generic(row, n),
            axis=1
        )
```

```python
import numpy as np

# List of ITIS component columns
itis_cols = [
    "cyc_dose1_ITIS",
    "cyc_dose2_ITIS",
    "cyc_dose3_ITIS",
    "cyc_dose4_ITIS",
    "cyc_dose5_ITIS",
    "cyc_dose6_ITIS",
    "cyc_dose7_ITIS",
    "cyc_dose9_ITIS",
    "cyc_dose10_ITIS",
    "cyc_dose11_ITIS",
]

def compute_cumulative_ITIS(row):
    values = row[itis_cols].dropna().values

    # If all ITIS_i are missing → cumulative result is NaN
    if len(values) == 0:
        return np.nan

    # Apply formula: 1 - Π(1 - ITIS_i)
    product = np.prod(1 - values)
    return 1 - product

# Create new column
cyc_merged_data["CYC_IV_ITIS"] = cyc_merged_data.apply(compute_cumulative_ITIS, axis=1)
```

```python
import numpy as np
import pandas as pd

def merge_visits_with_rtx(visits, mtp_wide):

    df = visits.copy()
    rtx = rtx_wide.copy()

    # Ensure Date_Of_Visit is datetime
    df["Date_Of_Visit"] = pd.to_datetime(df["Date_Of_Visit"], errors="coerce")

    # Merge wide IV data onto visits
    df = df.merge(rtx, on="RKD.ID", how="left")

    # Identify dose blocks
    dose_numbers = sorted(
        int(col.split("_")[1])
        for col in rtx.columns
        if col.startswith("rtx_") and col.endswith("_dose_date")
    )

    for n in dose_numbers:

        dose_date_col = f"rtx_{n}_dose_date"
        value_col     = f"rtx_{n}_dose_value"


        # Convert IV dates to datetime
        df[dose_date_col] = pd.to_datetime(df[dose_date_col], errors="coerce")

        # Condition:
        # Keep the dose if the visit occurs AFTER the dose date
        # Blank it otherwise
        mask = (
            df[dose_date_col].notna() &
            (df[dose_date_col] < df["Date_Of_Visit"])
        )

        # Blank outside window
        df.loc[~mask, [dose_date_col, value_col]] = [np.nan, np.nan]

    return df
```

```python
def update_rtx_doses(df, base_dose, target_dose, max_days=60):
    result = df.copy()

    base_date   = f"rtx_{base_dose}_dose_date"
    target_date = f"rtx_{target_dose}_dose_date"
    base_value  = f"rtx_{base_dose}_dose_value"
    target_value= f"rtx_{target_dose}_dose_value"

    # Parse dates
    result[base_date]   = pd.to_datetime(result[base_date], errors="coerce", dayfirst=True)
    result[target_date] = pd.to_datetime(result[target_date], errors="coerce", dayfirst=True)

    # Parse numeric values
    result[base_value]   = pd.to_numeric(result[base_value], errors="coerce")
    result[target_value] = pd.to_numeric(result[target_value], errors="coerce")

    def _per_patient(g):
        g = g.copy()

        d_base   = g[base_date].dropna()
        d_target = g[target_date].dropna()

        # Need both doses
        if d_base.empty or d_target.empty:
            return g

        base_date_val   = d_base.iloc[0]
        target_date_val = d_target.iloc[0]

        # Check interval
        interval_days = (target_date_val - base_date_val).days
        if pd.isna(interval_days) or interval_days > max_days:
            return g

        # Get dose values
        v_base   = g.loc[g[base_date].notna(), base_value].dropna()
        v_target = g.loc[g[target_date].notna(), target_value].dropna()

        base_val   = v_base.iloc[0]   if not v_base.empty   else np.nan
        target_val = v_target.iloc[0] if not v_target.empty else np.nan

        cumulative = base_val + target_val

        # Replace ALL non-missing base-dose rows
        mask = g[base_date].notna()
        g.loc[mask, base_date]  = target_date_val
        g.loc[mask, base_value] = cumulative

        # Blank target-dose after merge
        g[target_date]  = pd.NaT
        g[target_value] = np.nan

        return g

    return result.groupby("RKD.ID", group_keys=False).apply(_per_patient)
```

```python
def reorder_rtx_columns(df):

    cols = list(df.columns)

    # Identify dose numbers based ONLY on dose_date
    dose_numbers = sorted(
        int(c.split("_")[1])
        for c in cols
        if c.startswith("rtx_") and c.endswith("_dose_date")
    )

    # 1) Core columns (non-rtx)
    core_cols = [c for c in cols if not c.startswith("rtx_")]

    # 2) Dose blocks (ordered)
    dose_blocks = []
    for n in dose_numbers:
        block = [
            f"rtx_{n}_dose_date",
            f"rtx_{n}_dose_value",
            f"rtx_{n}_vanish_days",
            f"rtx_{n}_vanish_date",
        ]
        dose_blocks.extend([c for c in block if c in cols])

    # 3) Everything else starting with rtx_ (course columns, intervals, etc.)
    other_rtx_cols = [
        c for c in cols
        if c.startswith("rtx_") and c not in dose_blocks
    ]

    # Build final ordering
    ordered_cols = core_cols + dose_blocks + other_rtx_cols

    return df[ordered_cols]

rtx_merged_data = reorder_rtx_columns(rtx_merged_data)
rtx_merged_data 
```

```python
import numpy as np
import pandas as pd

# ---------------------------------------------
# 1. Sigmoid function (UNCHANGED)
# ---------------------------------------------
def sigmoid_curve(x, A, n, d):
    return A / (1 + np.exp(n * (x - d)))

# ---------------------------------------------
# 2. Interpolation functions for A(dose), d(dose), n(dose) (UNCHANGED)
# ---------------------------------------------
A1 = 0.70
A2 = 0.85
dose1 = 500
dose2 = 2000

m_A = (A2 - A1) / (dose2 - dose1)
b_A = A1 - m_A * dose1
def calculate_A(dose):
    return m_A * dose + b_A

d1 = 160
d2 = 200
m_d = (d2 - d1) / (dose2 - dose1)
b_d = d1 - m_d * dose1
def calculate_d(dose):
    return m_d * dose + b_d

# vanish logic (UNCHANGED)
def calculate_n_from_vanish(A, d, vanish_day):
    return np.log(99) / (vanish_day - d)

vanish1 = 240
vanish2 = 300
n1 = calculate_n_from_vanish(A1, d1, vanish1)
n2 = calculate_n_from_vanish(A2, d2, vanish2)

m_n = (n2 - n1) / (dose2 - dose1)
b_n = n1 - m_n * dose1
def calculate_n(dose):
    return m_n * dose + b_n

# ---------------------------------------------
# 3. ITIS computation for ALL doses 1–11 (equations unchanged)
# ---------------------------------------------
def compute_ITIS_generic(row, dose_num):
    dose_col = f"rtx_{dose_num}_dose_value"
    time_col = f"days_visit_minus_rtx{dose_num}"

    dose = row[dose_col]
    time = row[time_col]

    # Skip invalid values
    if pd.isna(dose) or pd.isna(time):
        return np.nan

    A = calculate_A(dose)
    d = calculate_d(dose)
    n = calculate_n(dose)

    return sigmoid_curve(time, A, n, d)

# Create ITIS columns for doses 1–11
for n in range(1, 29):
    dose_col = f"rtx_{n}_dose_value"
    time_col = f"days_visit_minus_rtx{n}"
    out_col  = f"rtx_dose{n}_ITIS"

    if dose_col in rtx_merged_data.columns and time_col in rtx_merged_data.columns:
        rtx_merged_data[out_col] = rtx_merged_data.apply(
            lambda row: compute_ITIS_generic(row, n),
            axis=1
        )
```

```python
import numpy as np

# List of ITIS component columns
itis_cols = [
    "CYC_Oral_ITIS",
    "CYC_IV_ITIS",
]

def compute_cumulative_ITIS(row):
    values = row[itis_cols].dropna().values

    # If all ITIS_i are missing → cumulative result is NaN
    if len(values) == 0:
        return np.nan

    # Apply formula: 1 - Π(1 - ITIS_i)
    product = np.prod(1 - values)
    return 1 - product

# Create new column
mtp_with_intervals["Cumulative__CYC_ITIS"] = mtp_with_intervals.apply(compute_cumulative_ITIS, axis=1)
```
